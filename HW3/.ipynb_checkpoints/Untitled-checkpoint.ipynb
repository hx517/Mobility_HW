```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 400, 0.1)
x
```

```python
y = 60*x - 0.15*x*x
```

```python
max(y)
```

```python
plt.plot(x,y)
plt.ylabel("flow q (vehicles/hour)")
plt.xlabel("density k (vehicles/mile)")
plt.title("Triangular Flow-Density Relation Diagram")
plt.show()
```
